```python
import tensorflow as tf
from tensorflow.keras.utils import plot_model
```

## Sequential API

```python
model=tf.keras.Sequential([
    tf.keras.layers.Dense(64,activation='relu',input_shape=(2,)),
    tf.keras.layers.Dense(16,activation='relu'),
    tf.keras.layers.Dense(1,activation='sigmoid')
])

# model = tf.keras.Sequential()

# model.add(tf.keras.layers.Dense(64, activation='relu', input_shape=(2,)))
# model.add(tf.keras.layers.Dense(16, activation='relu'))
# model.add(tf.keras.layers.Dense(1, activation='sigmoid'))
```

```python
plot_model(model,show_shapes='true')
```

### Functional API

single input and single output

```python
input=tf.keras.Input(shape=(2,))

hidden1=tf.keras.layers.Dense(64,activation='relu')(input)
hidden2=tf.keras.layers.Dense(16,activation='relu')(hidden1)
output=tf.keras.layers.Dense(1,activation='sigmoid')(hidden2)
```

```python
model=tf.keras.Model(inputs=input,outputs=output)
```

```python
plot_model(model,show_shapes='true')
```

Multple input and single output

```python
input1=tf.keras.Input(shape=(2,))
input2=tf.keras.Input(shape=(2,))

hidden1=tf.keras.layers.Dense(64,activation='relu')
hidden2=tf.keras.layers.Dense(16,activation='relu')

hidden1_out=hidden1(input1)
hidden2_out=hidden2(input2)

concat=tf.keras.layers.Concatenate()([hidden1_out,hidden2_out])

output=tf.keras.layers.Dense(1,activation='sigmoid')(concat)
```

```python
model=tf.keras.Model(inputs=[input1,input2],outputs=output)
```

```python
plot_model(model,show_shapes='true')
```

single input and multiple output

```python
input=tf.keras.Input(shape=(2,))

hidden1=tf.keras.layers.Dense(64,activation='relu')(input)
hidden2=tf.keras.layers.Dense(16,activation='relu')(hidden1)

output1=tf.keras.layers.Dense(1,activation='sigmoid')(hidden2)
output2=tf.keras.layers.Dense(1,activation='sigmoid')(hidden2)

model=tf.keras.Model(inputs=input,outputs=[output1,output2])
```

```python
model=tf.keras.Model(inputs=input,outputs=[output1,output2])
```

```python
plot_model(model,show_shapes='true')
```

multiple input and multiple outputs

```python
input1=tf.keras.Input(shape=(2,))
input2=tf.keras.Input(shape=(2,))

hidden1=tf.keras.layers.Dense(64,activation='relu')
hidden2=tf.keras.layers.Dense(16,activation='relu')

hidden1_out=hidden1(input1)
hidden2_out=hidden2(input2)

concat=tf.keras.layers.Concatenate()([hidden1_out,hidden2_out])

output1=tf.keras.layers.Dense(1,activation='sigmoid')(concat)
output2=tf.keras.layers.Dense(1,activation='sigmoid')(concat)
```

```python
model=tf.keras.Model(inputs=[input1,input2],outputs=[output1,output2])
```

```python
plot_model(model,show_shapes='true')
```
